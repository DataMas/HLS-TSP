```
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```
path=[]
with open('optimal.txt') as cfile:
     line = csv.reader(cfile, delimiter=',')
     for row in line:
       if row != '':
         path.append(row)

path=path[0][:-1]
```

```
coords=[]
with open('coords.txt') as cfile:
     line = csv.reader(cfile, delimiter=' ', skipinitialspace=True)
     for row in line:
       mini=[]
       if row != '':
        mini.append(float(row[0])) 
        mini.append(float(row[1])) 
        coords.append(mini)
```

```
names=[]
with open('names.txt') as cfile:
     line = csv.reader(cfile, delimiter=',')
     for row in line:
       if row != '':
        names.append(row)
```

```
ids=[i for i in range(128)]
```

```
dict_name = dict(zip(ids, names))
dict_cords = dict(zip(ids, coords))
```

```
G = nx.Graph()

for i in range(128):
  G.add_node(dict_name[i][0], pos=dict_cords[i])

for i in range(126):
  G.add_edge( dict_name[int(path[i])][0] , dict_name[int(path[i+1])][0] )

plt.figure(3,figsize=(15,15))
nx.draw(G, nx.get_node_attributes(G, 'pos'), with_labels=True, node_size=50)
```
